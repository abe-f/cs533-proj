```python
#  ../zsim/build/opt/zsim test.cfg

import h5py
import numpy as np
from readable_number import ReadableNumber
import pandas as pd


import os
from readable_number import ReadableNumber
import matplotlib.pyplot as plt
import seaborn as sns
rn = ReadableNumber
sns.set_theme()

f = h5py.File("zsim.h5", 'r')

print(f.keys())

dset = f["stats"]["root"]

num_private_sets = np.sum(dset['l3']['profNumReclaimedSets'], 1)
num_private_sets = np.divide(num_private_sets, 1048576)
phase = 100000/1000000 # x is in millions
x_vals = np.arange(0, num_private_sets.size*phase, phase)

fig, ax = plt.subplots()
ax.plot(x_vals, num_private_sets);

fig.tight_layout()

ax.set_ylabel("Fraction of Sets with >=1 Shared Lines")
ax.set_xlabel("Cycles (millions)")
```
